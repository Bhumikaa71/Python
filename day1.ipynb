```python
# # Question 1: Copying a File with Error Handling
# Question:
# Write a Python program that reads a text file named input.txt and copies its contents
#  to a new file named output.txt. Handle the case where the input file is missing and print an appropriate error message.

# Question 1: Copying a File with Error Handling
# Question:
# Write a Python program that reads a text file named input.txt and copies its contents
# to a new file named output.txt. Handle the case where the input file is missing.

try:
    with open('input.txt', 'r') as infile:
        content = infile.read()

    with open('output.txt', 'w') as outfile:
        outfile.write(content)

    print("File copied successfully.")

except FileNotFoundError:
    print("Error: The file 'input.txt' was not found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")


```

# Transpose the scraatch

```python
import numpy as np

# Original matrix
matrix1 = np.array([[2, 4], [5, 7]])

# Get number of rows and columns
rows = len(matrix1)
cols = len(matrix1[1])

# Create an empty matrix with swapped dimensions
transpose = [[0 for _ in range(rows)] for _ in range(cols)]

# Fill the transpose manually
for i in range(rows):
    for j in range(cols):
        transpose[j][i] = matrix1[i][j]

transpose = np.array(transpose)
print("Original matrix:")
print(matrix1)

print("\nTranspose (from scratch):")
print(transpose)
```

## Exercise 6 : Matrix Inverse

```python
import numpy as np

matrix1 = np.array([[2, 4], [5, 7]])

# Calculate inverse
inverse = np.linalg.inv(matrix1)
print("Inverse of matrix1:")
print(inverse)

# Multiply matrix by its inverse
result = np.dot(matrix1, inverse)
print("\nmatrix1 * inverse:")
print(result)
```
